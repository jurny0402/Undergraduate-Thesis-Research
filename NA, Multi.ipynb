```r
combined_raw <- combineTCR(contig_list,
                      samples = c("FT1","FT2","FT3","FT4","FT5"),
                      cells = "T-AB",
                      ID = c("a","a","a","c","c"))
```

```r
quantContig(combined_raw,
           cloneCall = "strict",
           chain = "both",
           scale = T)
quantContig(combined_raw,
           cloneCall = "strict",
           chain = "both",
           scale = T,
           exportTable = T)
```

```r
combined <- combineTCR(contig_list,
                      samples = c("FT1","FT2","FT3","FT4","FT5"),
                      cells = "T-AB",
                      ID = c("a","a","a","c","c"),
                      removeNA = T,
                      removeMulti = T)
```

```r
quantContig(combined,
           cloneCall = "strict",
           chain = "both",
           scale = T)
quantContig(combined,
           cloneCall = "strict",
           chain = "both",
           scale = T,
           exportTable = T)
```

```r
combined_NA <- combineTCR(contig_list,
                      samples = c("FT1","FT2","FT3","FT4","FT5"),
                      cells = "T-AB",
                      ID = c("a","a","a","c","c"),
                      removeMulti = T)
```

```r
quantContig(combined_NA,
           cloneCall = "strict",
           chain = "both",
           scale = T)
quantContig(combined_NA,
           cloneCall = "strict",
           chain = "both",
           scale = T,
           exportTable = T)
```

```r
combined_mul <- combineTCR(contig_list,
                      samples = c("FT1","FT2","FT3","FT4","FT5"),
                      cells = "T-AB",
                      ID = c("a","a","a","c","c"),
                          removeNA = T)
```

```r
quantContig(combined_mul,
           cloneCall = "strict",
           chain = "both",
           scale = T)
quantContig(combined_mul,
           cloneCall = "strict",
           chain = "both",
           scale = T,
           exportTable = T)
```
